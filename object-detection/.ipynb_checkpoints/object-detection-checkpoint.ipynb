```python
import torch
from matplotlib import pyplot as plt
import cv2
```

## Bounding Boxes

```python
img = plt.imread("./img/catdog.jpeg")
plt.imshow(img);
```

```python
# bounding box upper right corner and lower left corner (x,y) 
dog_bbox, cat_bbox = [600.0, 95.0, 1050.0, 600.0], [300.0, 220.0, 640.0, 600.0]
```

```python
#@save
def bbox_to_rect(bbox, color):
    """Convert bounding box to matplotlib format."""
    # Convert the bounding box (upper-left x, upper-left y, lower-right x,
    # lower-right y) format to the matplotlib format: ((upper-left x,
    # upper-left y), width, height)
    return plt.Rectangle(
        xy=(bbox[0], bbox[1]), width=bbox[2]-bbox[0], height=bbox[3]-bbox[1],
        fill=False, edgecolor=color, linewidth=2)
```

```python
fig = plt.imshow(img)
fig.axes.add_patch(bbox_to_rect(cat_bbox, "blue"))
fig.axes.add_patch(bbox_to_rect(dog_bbox, "red"))
```
